```python
#import necessary packages 
import numpy as np
import pandas as pd 
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#import adjacency matrix 
df = pd.read_csv('Alibaba-network.csv',index_col=0)
```

```python
#convert adjacency matrix to edgelist dataframe directional
def edgelist(df):
    a = df.values
    c = df.columns
    n = len(c)
    
    c_ar = np.array(c)
    out = np.empty((n, n, 2), dtype=c_ar.dtype)
    
    out[...,0] = c_ar[:,None]
    out[...,1] = c_ar
    
    mask = ~np.eye(n,dtype=bool)
    df_out = pd.DataFrame(out[mask], columns=[['Source','Target']])
    df_out['Weight'] = a[mask]
    return df_out
```

```python
## create an edge list with all the connections 
edge_df = pd.DataFrame()
edge_df=edgelist(df)
edge_df.columns = edge_df.columns.get_level_values(0)
edge_df = edge_df.query('Weight != 0.0')
edge_df=edge_df[['Source','Target']]
edge_tuple = edge_df.to_records(index=False,column_dtypes = {"Source":"int32", "Target":"int32"})
edges = list(edge_tuple)
```

```python
#create node list 
nodes = list(df.index)
```

```python
##create an empty graph 
G = nx.DiGraph()
```

```python
##load the edges and nodes 
G.add_edges_from(edges)
G.add_nodes_from(nodes)
```

```python
remove = [node for node,degree in dict(G.degree()).items() if degree < 4]
G.remove_nodes_from(remove)
```

```python
remove = [node for node,degree in dict(G.degree()).items() if degree < 2]
G.remove_nodes_from(remove)
```

```python
# visualize the network -  betweenness 
from pylab import rcParams
rcParams['figure.figsize'] = 14, 10
pos = nx.spring_layout(G, scale=20, k=3/np.sqrt(G.order()))
betCent = nx.betweenness_centrality(G, normalized=True, endpoints=True)

high_light_nodes = {4, 78, 293, 162, 1}
node_color = [betCent[v ]* 20000 if v in high_light_nodes else 0 for v in G ]
node_size = [betCent[v ]* 10000 if v in high_light_nodes else 0 for v in G ] 
#node_size = [v * 10000 if k in high_light_nodes else 10 for k,v in zip(G,betCent.values())]

nx.draw(G, pos, node_color=node_color, 
        with_labels=True, 
        node_size=node_size,
        font_size=10,
        linewidths=1)
```

```python
# visualize the network - degree
from pylab import rcParams
rcParams['figure.figsize'] = 14, 10
pos = nx.spring_layout(G, scale=20, k=3/np.sqrt(G.order()))
betCent = nx.betweenness_centrality(G, normalized=True, endpoints=True)

high_light_nodes = {1, 78, 521, 293, 284}
node_color = [20000.0 * G.in_degree(v) if v in high_light_nodes else 0 for v in G]
node_size = [100.0 * G.in_degree(v) if v in high_light_nodes else 10 for v in G] 
#node_size = [v * 10000 if k in high_light_nodes else 10 for k,v in zip(G,betCent.values())]

nx.draw(G, pos, node_color=node_color, 
        with_labels=True, 
        node_size=node_size,
        font_size=12,
        linewidths=1)
```

```python
# visualize the network - closeness

rcParams['figure.figsize'] = 14, 10
pos = nx.spring_layout(G, scale=20, k=3/np.sqrt(G.order()))
cloCent = nx.closeness_centrality(G)

high_light_nodes = {1, 127, 92, 364, 78}
node_color = [20000.0 * cloCent[v] if v in high_light_nodes else 0 for v in G]
node_size = [10000.0 * cloCent[v] if v in high_light_nodes else 10 for v in G] 

nx.draw(G, pos, node_color=node_color, 
        with_labels=True, 
        node_size=node_size,
        font_size=12,
        linewidths=1)

```

```python
##sort by betweenness centrality 
##sorted(betCent, key=betCent.get, reverse=True)[:5]
sorted(betCent.items(), key = lambda x: x[1], reverse= True)[:5]
```

```python
##sort by in degree centrality
node_in_degree = nx.in_degree_centrality(G)
##sorted(node_in_degree, key=node_in_degree.get, reverse=True)[:5]
sorted(node_in_degree.items(), key = lambda x: x[1], reverse= True)[:5]
```

```python
##sort by closeness centrality 
cloCent = nx.closeness_centrality(G)
##sorted(cloCent, key=cloCent.get, reverse=True)[:5]
sorted(cloCent.items(), key = lambda x: x[1], reverse= True)[:5]
```
